```python
# Import numpy library
import numpy as np
```

```python
# Create my matrices from the homework
x = np.array([[-1,0,3],[2,-4, 1],[ 3,4,-2]])
y = np.array([[ 2,1,0],[0,-2,-4],[-1,3,8]])
print('x = \n',x)
print('y = \n',y)
print("time")
```

```python
# Matrix addition
sumMatrix = x+y
print('x+y =\n',sumMatrix)
```

```python
# Matrix multiply. Notice this is @ and not *
multMatrix = x@y
print('x*y = \n',multMatrix)

# Element-wise multiplication
multElementsMatrix = x*y
print('x*y (element by element) = \n',multElementsMatrix)
```

```python
# Now, we will see how to use these in the context of computer graphics.
from PIL import Image
```

```python
# import the image
img = Image.open("monaLisa.png")
```

```python
# Here is one way to visualize the image
img.show()
```

```python
# And, here is another 
display(img)
```

```python
#Get basic details about the image
print(img.format)
print(img.mode)
print(img.size)
```

```python
# Convert the image to Grayscale 
imgGray = img.convert('L')
display(imgGray)
```

```python
# Convert the image to a matrix
imageArray = np.asarray(imgGray)
# Lighten the image
alteredImage = imageArray+20
# Convert back to an image 
img2 = Image.fromarray(alteredImage)
# Display the new image 
display(img2)
```

```python
# Invert the image
alteredImage = 255 - imageArray
# Convert back to an image 
img2 = Image.fromarray(alteredImage)
# Display the new image 
display(img2)
```

```python
# For fun let's play with this idea a bit. 
alteredImage = np.copy(imageArray)
borderSize = 90
alteredImage[borderSize:-borderSize,borderSize:-borderSize] = 255 - alteredImage[borderSize:-borderSize,borderSize:-borderSize]
# Convert back to an image 
img2 = Image.fromarray(alteredImage)
display(img2)
```
