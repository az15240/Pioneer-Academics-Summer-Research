# Invert an image in submatrices

Take random submatrices and invert them successively in an image. 

Created by Tim Chartier

```python
import numpy as np
from PIL import Image
from random import randint

# import the image
img = Image.open("monaLisa.png")
# Convert the image to Grayscale 
imgGray = img.convert('L')

# retrieve size of image
width, height = img.size

# Convert the image to a matrix
imageArray = np.asarray(imgGray)

submatrixSize = 75
numberOfSwappedSubmatrices = 20

alteredImage = np.copy(imageArray)
for _ in range(numberOfSwappedSubmatrices):
    rowIndex = randint(0, height-submatrixSize-1)
    colIndex = randint(0, width-submatrixSize-1)
    alteredImage[rowIndex:rowIndex+submatrixSize,colIndex:colIndex+submatrixSize] = 255 - alteredImage[rowIndex:rowIndex+submatrixSize,colIndex:colIndex+submatrixSize]

# Convert back to an image 
display(Image.fromarray(alteredImage))
```
